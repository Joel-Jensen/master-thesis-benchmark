```python
import matplotlib.pyplot as plt
import numpy as np

# Sample raw data (per-query times in ms)
queries = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8']
system_pg_mooncake = [789,1339,1215,1036,925,1804,914,2691]
system_timescale = [1708,295,2238,521,268,4911,3307,5911]
system_postgres_brin = [27247,916,2617,43,1865,5444,5898,25769]
system_postgres_partition = [7766,788,3242,3,1775,5532,5484,9834]
system_clickhouse = [168,265,631,279,378,1201,348,2235]

# Cumulative sum
system_pg_mooncake_cumulative = np.cumsum(system_pg_mooncake)
system_timescale_cumulative = np.cumsum(system_timescale)
system_postgres_brin_cumulative = np.cumsum(system_postgres_brin)
system_postgres_partition_cumulative = np.cumsum(system_postgres_partition)
system_clickhouse_cumulative = np.cumsum(system_clickhouse)

# Plot
plt.figure(figsize=(9, 6))

plt.plot(queries, system_pg_mooncake_cumulative, marker='o', linestyle='-', label='pg_mooncake')
plt.plot(queries, system_timescale_cumulative, marker='s', linestyle='--', label='TimescaleDB')
plt.plot(queries, system_postgres_partition_cumulative, marker='^', linestyle=':', label='PostgreSQL Partition')
#plt.plot(queries, system_postgres_brin_cumulative, marker='^', linestyle=':', label='PostgreSQL BRIN')
plt.plot(queries, system_clickhouse_cumulative, marker='*', linestyle=':', label='ClickHouse')
# Axis labels
plt.xlabel('Query')
plt.ylabel('Cumulative Time (ms)')

# Style
plt.grid(True, linestyle=':', linewidth=0.5)
plt.title('Query Execution 1 year')
plt.legend()
plt.tight_layout()

# IEEE-style font
plt.rcParams.update({'font.family': 'serif', 'font.size': 10})

# Save to file
#plt.savefig('cumulative_query_times.pdf', format='pdf')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample raw data (per-query times in ms)
queries = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8']
system_pg_mooncake = [789,1339,1215,1036,925,1804,914,2691]
system_timescale = [1708,295,2238,521,268,4911,3307,5911]
system_postgres_brin = [27247,916,2617,43,1865,5444,5898,25769]
system_postgres_partition = [7766,788,3242,3,1775,5532,5484,9834]
system_clickhouse = [168,265,631,279,378,1201,348,2235]

# Create figure and axis
plt.figure(figsize=(8, 6))

# Create stacked bar chart
x = np.arange(len(queries))  # the label locations
width = 0.35  # the width of the bars

# Plot the bars
plt.bar(x - width/2, system_timescale, width, label='TimescaleDB')
plt.bar(x + width/4, system_postgres_partition, width, label='PostgreSQL Partition')
plt.bar(x - width/4, system_pg_mooncake, width, label='pg_mooncake')
plt.bar(x - width/2, system_clickhouse, width, label='ClickHouse')
#plt.bar(x + width/4, system_postgres_brin, width, label='PostgreSQL BRIN')


# Customize the plot
plt.xlabel('Query')
plt.ylabel('Time (ms)')
plt.title('Query Execution 1 year')
plt.xticks(x, queries)
plt.legend()

# Style
plt.grid(True, linestyle=':', linewidth=0.5, axis='y')
plt.tight_layout()

# IEEE-style font
plt.rcParams.update({'font.family': 'Times New Roman', 'font.size': 10})

# Save to file
#plt.savefig('query_times_stacked.pdf', format='pdf')

plt.show()
```
